 ### Problem 2
 (40 points) A ball is kicked
at the same time as a container, initially at rest, starts sliding down a friction less incline plane, as seen in the figure. The kick provides an initial velocity to the ball
$v_{0,b}= v_{0,b}^x \hat{i} + v_{0,b}^x \hat{j} $.
After a time t the ball falls inside the container, as shown in the picture. 
All the relevant physical dimensions to solve the problem are provided (H,A,D).
In order to solve this problem take the origin of coordinates where the ball starts, $r_{0,b}=(0,0)$.
Take the positive motion direction up and to the left, for the ball.
For the container, the motion will be along the surface of the incline. The function that describes the displacement of the container along the surface of the plane is $L_{c}=\frac{1}{2}a_ct^2$.


* Ball's initial velocity along x $v_{0,b}^x$ m/s
* Ball's initial velocity along y $v_{0,b}^y$ m/s
* Mass of the Container m = 2 kg.
* H = 10 m
* A = 10 m
* D = 50 m
* g=9.8m/s$^2$


* *A* (5 points) What is the angle ($\theta$) of the incline?

 $\theta$=arctan(H/A)=68.8$^o$

* *B* (5 points) What is the acceleration $a_c$ of the container along the surface of the plane?

 a =g sin $\theta$ =9.1 m/s$^2$

* *C* (5 points) What is the equation for the magnitude of the displacement of the container $L_c(t)$ along the plane? 

  L$_c$ =$\frac{1}{2}$g sin$\theta$ t$^2$

* *D* (5 points) What is the equation for the horizontal displacement of the ball, $x_b(t)$ in the air? Write this equation as a vector.

    $x_{b}(t)=v_{0,b}^xt\hat{i}$

* *E* (5 points) What is the equation for the vertical displacement of the ball, $y_b(t)$ in the air? Write this equation as a vector.

    $y_{b}(t)=v_{0,b}^y t -\frac{1}{2}gt^2\hat{j}$
    
* F (10 points) How long (time t) does it take from the initial kick for the ball to fall inside the bucket ( time of collision).    

  $L_{c,x}(t)+x_b(t)=A+D$
  $\frac{1}{2}$g sin$\theta$ cos$\theta$ t$^2 + v_{0,b}^xt =A + D $
  See operations bellow.
  
* G  (5 points) How much work does gravity  on the ball in the air?
 
 We use the work energy principle. The net work done by gravity is the net change in kinetic energy of the ball.
 
 $W_g=\Delta E_{kin}$
 
 $E_{k,t=0}=1/2mv_0^2=\frac{1}{2}m_b(v_{0,x}^2+v_{0,y}^2)$
 
 $E_{k,t}=1/2mv_1^2=\frac{1}{2}m_b(v_{t,x}^2+v_{t,y}^2)$
 
 Using $v_{t,x}=v_{0,x}$ and $v_{t,y}=v_{0,y}-gt$ and t=2.5 s
 
 $\Delta E_k = -94.9$ J (see below)  
  

```python
V0x = 15 #m/3
V0y = 20 # m/s
H = 46.01 #m
A = 17.81 #m
D = 30 # m
g = 9.8 #m/s
m = 5 #kg
L = np.sqrt(A**2+H**2)
theta = np.arccos(A/L)*180/np.pi
a = g* (H/L)
t = (-V0x+np.sqrt(V0x**2+2*(A/L)*a*(A+D)))/(A/L)/a

print("L,t,a,theta=",L,t,a,theta )

s = 0.5*g*t**2
print("s = ",s)
m=0.5 #kg
vx = V0x
vy = (V0y - g*t)
print("vx,vy =", vx,vy)
Ek0= 0.5*m*(V0x**2+V0y**2)
Ek1= 0.5*m*(vx**2+vy**2)
print("delta Ek = ", Ek1-Ek0)

a = 0.5*g*((H**2+H*A)/L**2 - 1)
b = V0x+V0y
c = A+D+H
print(4*a*c,b**2,b)
t = (-b+np.sqrt(b**2+4*a*c))/(2*a)
print("time= ",t)
t = 2.5
h = V0y*t-0.5*g*t**2
x = V0x*t
print(x,h)
l = x-30
theta = np.arctan(h/l)
s = 0.5*g*np.sin(theta)*t**2
sy = s*np.sin(theta)
sx = s*np.cos(theta)
print(s,sy,sx,sy+h,sx+l)
```
