### Problem 1

I want to raise a $1\,\mathrm{kg}$ object from the ground by pulling on a rope that passes over a pulley which has mass $200\,\mathrm{g}$ and radius $10\,\mathrm{cm}$ and can be treated as frictionless.  For a solid disk the moment of inertia is $\frac{1}{2}MR^{2}$.The 1 kg object is initially at rest and I would like to lift it at a constant speed of $1\,\mathrm{ms^{-1}}$. 

(a) (5 points ) In order to have the weight be moving with this velocity $0.5\,\mathrm{s}$ after I start pulling on the rope, what should the force I apply be. (Assume uniform acceleration of the object)?

v = at
1=a×0.5 a=2ms$^{−2}$

$Fr−Tr=1/2Mr^2\alpha \rightarrow F−T=1/2Ma = 9.11\times10^6 m$

$ T−mg=ma\rightarrow T=mg+ma$

$F= 1/2 Ma+mg+ma=12N$


(b) (5 points) What is the angular acceleration of the pulley while I am doing this?

$\alpha = a/r = 2/0.1 =20rads^{−2}$

(c) (5 points) Once the object has reached a constant speed of $1\,\mathrm{ms^{-1}}$ what force should I apply to maintain that speed.

At constant speed a=0 so F=mg=1×9.8=9.8N

(d) (5 points) What is the angular velocity (in rpm) of the pulley while I am raising the object with a constant speed  of $1\,\mathrm{ms^{-1}}$?

$\omega= v/r = 1/0.1 = 10rads^{−1} = 10× 60/(2\pi) =95.5 rpm$

(e) (5 points) What is the angular momentum of the pulley while I am raising the object with a constant speed  of $1\,\mathrm{ms^{-1}}$? Give both the magnitude and the direction (ie. left, right, up, down, in to the page, out of the page).

$L=I\omega = 1/2 M r^2\omega=0.5×0.2×0.1^2 ×10=0.01kgm2s^{−1}$

out of page by righthandrule

### Problem 2 
A 100g solid ball of radius 2 cm on a string is raised to an angle of 50$^{o}$ and then released. The distance from the point at which the string is attached to the center of the ball is 1m. The ball swings down and hits an identical ball. 


* a) How much kinetic energy does the ball on the string have just before it hits the second ball?
   
    $\Delta h= 1-\cos50^{o}=0.357\mathrm{m}$
    
    $KE=-mg\Delta h=0.1\times 9.8\times0.357=0.35 \mathrm{J}$

* b) The collision is elastic. How much kinetic energy does the second ball have just after the collision?

    All of the kinetic energy is transferred to the second ball so

    $KE=0.35\mathrm{J}$

* c) The second ball then rolls up a 15$^{o}$ incline to a height of 10 cm. What is the magnitude of the velocity at the top of the slope.

    KE at top of ramp is $0.35-0.1\times9.8\times0.1=0.252\mathrm{J}$

    For a rolling ball $KE=\frac{1}{2}m(1+\frac{2}{5})v^{2}$ so

    $v^{2}=\frac{0.252\times 2\times 5}{0.1\times7}=3.6$

    $v=1.9\mathrm{m\,s^{-1}}$

* d)  What is the angular momentum of the ball at this point? Give  magnitude and direction (ie. left, right, up, down, in to page, out of page).

    $L=I\omega$

    $I=\frac{2}{5}mr^{2}=\frac{2}{5}\times0.1\times.02^{2}=0.000016\mathrm{kg\,m^{2}}$

    $\omega=\frac{v}{r}=\frac{1.9}{0.02}=95\mathrm{s^{-1}}$

    $L=0.00152\mathrm{kg\,m^{2}\,s^{-1}}$ in to the page

* e) The ball then flies off the ramp. At what distance $x$ from the bottom of the ramp does it strike the ground? 

    First we need to find the time it takes until the ball hits the ground

    $ y = y_0 +v_0t -  \frac{1}{2}gt^{2}$   
    
    $0=0.1+1.9\sin 15^{o}t-\frac{1}{2}gt^{2}$

    $4.9t^{2}-0.49t-0.1=0$

    $t=0.2 \mathrm{s}$

    $x=1.9\times\cos 15^{o}t$

    $x=0.37\mathrm{m}$

### Problem 3

A robot iceskater is initially at rest and then fires rockets on the end of it's arms, each of which exerts a force of 100N for 0.5s. Once the rockets have turned off and the robot has reached a constant angular velocity, the robot raises its arms, leading to a change in its angular velocity.


* a) Find the moment of inertia of the robot when it's arms are extended considering that the rockets weigh 2 kg each, each arm weighs 3 kg, the body of the robot is a 50 kg cylinder with radius 10cm. The arms have length 1.1m and are attached to the edge of the cylinder. The rockets may be considered as point masses. You may neglect the moment of inertia of the robot's head.

    $ I = 2I_a +2I_R +I_c = 2m_a(\frac{l^2}{12}+(R+l/2)^2)+2m_R(l+R)^2+1/2MR^2$
    
    $I=2\times2\times1.2^{2}+2\times3\times(\frac{1}{12}\times1.1^{2}+0.65^{2})+\frac{1}{2}\times50\times0.1^2$

    $I=5.76+3.14+0.25=9.15\,\mathrm{kg\,m^{2}}$

* b) What is the angular acceleration of the robot during the 0.5s that the rockets are on?
    
    $\sum \tau =2 \times 100 \times 1.2 =240\,\mathrm{N\,m}$

    $\alpha=\frac{\sum \tau}{I}=\frac{240}{9.67}=26.23\,\mathrm{s^{-2}}$

* c) What is the angular velocity achieved when the rockets have been turned off but the robot has not yet raised it's arms? Give your answer in both s$^{-1}$ and rpm. 

    $\omega=\alpha t\rightarrow\omega=26.23\times0.5=13.115\mathrm{s^{-1}}=125.24\mathrm{rpm}$

* d) What is the angular momentum when the rockets have been turned off but the robot has not yet raised it's arms? Give magnitude and direction (ie. left, right, up, down, in to page, out of page)? 
    $L=I\omega=9.15\times13.115=120\,\mathrm{kg\,m^{2}\,s^{-1}}$, up

* e) What is the moment of inertia of the robot after it has raised it's arms? 
    $ I=I_c+2I_{arms,vertical}+I_{rocket,vertical}= 1/2MR^2 +2m_aR^2+2m_RR^2$
    
    $I=2\times5\times0.1^{2}+\frac{1}{2}\times50\times0.1^{2}=0.35\,\mathrm{kg\,m^{2}}$

* f) What is the angular velocity of the robot once it has raised it's arms? Give your answer in both s$^{-1}$ and rpm. 
    
    Angular momentum is conserved so 

    $I_{extended}\omega_{i}=I_{raised}\omega_{f}\rightarrow\omega=\frac{120}{0.35}=342.9\mathrm{s^{-1}}=3274\mathrm{rpm}$

* g) What is the magnitude of the linear velocity of one of the rockets once the robot has raised it's arms?

    $v=\omega r=342.9\times0.1=34.29\,\mathrm{m\,s^{-1}}$
